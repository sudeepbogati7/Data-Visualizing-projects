## How many job title were represented by only one person in 2013 ? (Eg. Job Titles with only one occurence in 2013 )

```python

jobs_in_2013 = df.loc[(df.Year == 2013)] 
jobs_in_2013[['JobTitle','Year']]
unique_jobs_in_2013 = jobs_in_2013['JobTitle'].unique()

count = 0
for i in unique_jobs_in_2013:
    count +=1
    print(i,count)
    
print('\n')
print("There are ",count, "unique jobs in 2013 ! ")
    
no_of_jobs = np.size(unique_jobs_in_2013)
```

### How many people have the word Chief in their JobTitle?

```python
def chief_string(title):
    if 'chief' in title.lower().split():
        return True
    else:
        return False
```

```python
sum(sal['JobTitle'].apply(lambda x : chief_string(x)))
```

```python
#alteratively :

sal[sal['JobTitle'].apply(lambda x : chief_string(x))]['JobTitle'].count()
```
