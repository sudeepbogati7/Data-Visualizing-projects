```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df3 = pd.read_csv('df3')
df3.head()
```

##### Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion. Remember back to your matplotlib lecture...

```python
plt.figure(figsize =(10,3))
plt.scatter(x=df3['b'],y=df3['a'], color='red',marker = 'o', linewidth = 1, edgecolors = 'black')
plt.xlabel('values of b')
plt.ylabel('values of a')
plt.xlim(-0.2,1.3)
plt.ylim(-0.2,1.3)
plt.show()
```

##### Create a histogram of the 'a' column.

```python
plt.figure(figsize =(10,5))
sns.histplot(x=df3['a'], data= df3, color ='b', bins = 30, alpha = 0.4, edgecolor = 'black')
plt.grid(True, alpha = 0.5)
plt.fill_between(df3['a'] ,df3['a'],color='gray',alpha =0.5)
plt.ylabel('Frequency')
plt.xlabel("values of a")
plt.show()
```

##### Create a boxplot comparing the a and b columns

```python
boxes = plt.boxplot([df3['a'], df3['b']],labels = ['a','b'],whiskerprops = {'color':'red', 'linewidth':1, 'linestyle' : '--' }, patch_artist = True,medianprops = {'linewidth':1.5, 'color':'blue'} )
for box in boxes['boxes']:
    box.set(color = 'red',linewidth = 0.5, facecolor = 'lightgray')
plt.show() 
```

#####  Create a kde plot of the 'd' column 

```python
plt.figure(figsize = (6,4))
sns.kdeplot(x = df3['d'], data = df3 ,color = 'red')
plt.grid(True)
```
